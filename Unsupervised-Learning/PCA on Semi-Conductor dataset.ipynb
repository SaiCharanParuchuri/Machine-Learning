```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/secom/secom.data"
,sep=" ",header=None)
```

```python
data.head()
```

```python
a=list(data.isna().sum())
a.sort()
print(a[len(a)-10:len(a)])
```

```python
for i in list(data):
    data[i]=data[i].fillna(data[i].mean())
data.head()
```

```python
def mean(d):
    return d-np.mean(d)
data=data.apply(mean)
data.head()
```

<b>Standardizing Data<b>

```python
for i in list(data):
  mean = data[i].mean()
  std = data[i].std()
  if std>0:
    data[i] = data[i].apply(lambda x:(x-mean)/std)    
data.shape
```

```python
from sklearn.preprocessing import StandardScaler
std_data=StandardScaler().fit_transform(data)
std_data.shape
```

<b>Finding Covariance Martix<b>

```python
covar_matrix = np.matmul(std_data.T,std_data)
covar_matrix.shape
```

<b>Finding Eigen Values and Eigen vectors<b>

```python
from scipy.linalg import eigh
values,vectors=eigh(covar_matrix)
print(values.shape)
print(vectors.shape)
```

Finding variance for each principal components

```python
var=values/np.sum(values)*100
components= list(range(len(var)))
cumulative_var = np.cumsum(var)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
plt.xlabel("Components")
plt.ylabel("cumulative variance")
plt.title("Components VS cumulative variance")
plt.bar(components,var,align='edge')
plt.show()
```
